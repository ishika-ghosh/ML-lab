### Assignment 1: Write a Python program to create a plot to get a general Statistics of Iris data. (Using the dataset “iris.csv”).

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('iris')
df.describe().plot(kind='area')
plt.show()
```

### Assignment 2:Write a Python program to create a graph to find relationship between the sepal length and width. (Using the dataset “iris.csv”).

```python
df=sns.load_dataset('iris')
sns.scatterplot(x='sepal_length', y='sepal_width',data=df)
plt.show()
```

### Assignment 3: Write a Python program to create a join plot using "hexbin" to describe individual distributions on the same plot between Sepal length and Sepal width. (Using the dataset“iris.csv”).

```python
df = sns.load_dataset('iris')
sns.jointplot(x="sepal_length",y="sepal_width",kind="hex",data=df)
plt.show()
```

### Assignment 4:Write a Python program to create a joinplot and add regression and kernel density fits using "reg" to describe individual distributions on the same plot between Sepal length and Sepal width. (Using the dataset “iris.csv”).

```python
df = sns.load_dataset('iris')
sns.jointplot(x="sepal_length",y="sepal_width",kind="reg",data=df)
plt.show()
```

### Assignmet 5: Write a Python program to find the correlation between variables of iris data. Also create a heatmap using Seaborn to present their relations (Using the dataset “iris.csv”).

```python
df = sns.load_dataset("iris")
x = df.iloc[:, :4]
corr = x.corr()
sns.heatmap(corr, cmap=sns.diverging_palette(220, 10, as_cmap=True), linewidths=1)
plt.show()
```
